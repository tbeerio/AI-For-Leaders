# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

### **My Answers to the Weekly Graph Question: Using Preattentive Attributes**

To direct the viewer’s attention to the last three months of data, we can apply **preattentive attributes** such as **color, size, position, and annotations**. Here are several ways to emphasize the final three months in our graphs:

1. **Color Contrast** – Highlight the last three months in **a different color** while keeping the rest of the line muted (e.g., gray).
2. **Line Thickness & Style** – Make the last three months **bolder or thicker** while using a lighter or dashed style for previous months.
3. **Annotations** – Add **labels or markers** at key points in the last three months to draw attention.
4. **Background Shading** – Shade the final three months with a subtle background highlight to differentiate them.
5. **Zoomed Inset** – Create a **zoomed inset** of the last three months in a smaller subplot to focus attention.
6. **Animation or Step Highlighting** – If interactive, progressively reveal data, showing earlier months faded out.

By using these techniques, I think that we can ensure that the audience’s **eye is naturally drawn** to the most relevant part of the data first.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
# 4) Storytelling With Data - Figure 4.8 (Page 111)

import matplotlib.pyplot as plt

# Sample data for replication
concerns = [
    "Engine power is less than expected", "Tires make excessive noise while driving",
    "Engine makes abnormal/excessive noise", "Seat material concerns",
    "Excessive wind noise", "Hesitation or delay when shifting",
    "Bluetooth system has poor sound quality", "Steering system/wheel has too much play",
    "Bluetooth system is difficult to use", "Front seat audio/navigation controls"
]
values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]


colors = ["#b04a44" if val >= 10 else "#808080" for val in values]

# Create horizontal bar chart
plt.figure(figsize=(8, 6))
plt.barh(concerns[::-1], values[::-1], color=colors)  # Reverse order for correct display
plt.xlabel("Concerns per 1,000")
plt.title("Top 10 Design Concerns")

# Add values to bars
for index, value in enumerate(values[::-1]):
    plt.text(value + 0.2, index, f"{value:.1f}", va="center", fontsize=10)

# Remove chart border
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```
