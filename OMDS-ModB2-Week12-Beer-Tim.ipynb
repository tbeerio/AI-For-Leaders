# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

### Weekly Graph Question: My Answer

In the first visualization, each week's data is shown in separate side-by-side plots. This format helps isolate each week's pattern, making it easier to compare internal trends such as peaks and troughs within a given week. However, it can be more difficult to directly compare the two weeks side by side, since the reader needs to mentally align the patterns across separate axes.

The second visualization overlays both weeks on the same graph. This allows for immediate and direct comparison between the two weeks and makes it easier to spot differences or similarities in daily visitor patterns. However, this approach may become cluttered if more weeks or categories are added, potentially reducing clarity.

Each method has its advantages. The side-by-side plots support week-specific analysis, while the combined view enhances cross-week comparisons. The choice depends on the goal of the analysis and the complexity of the dataset.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import pandas as pd

years = list(range(2008, 2015))
data = {
    'A': [400, 420, 425, 400, 300, 270, 260],
    'B': [360, 400, 410, 385, 290, 265, 250],
    'C': [0, 0, 100, 180, 200, 230, 210],
    'D': [0, 0, 0, 170, 190, 180, 200],
    'E': [0, 0, 0, 0, 0, 100, 180]
}

df = pd.DataFrame(data, index=years)

# Plot
plt.figure(figsize=(10, 6))
for col in df.columns:
    plt.plot(df.index, df[col], color='gray', linewidth=2, alpha=0.7)
    plt.text(df.index[0] - 0.3, df[col].iloc[0], col, va='center', ha='right', fontsize=10, color='gray')

# Highlight last data points
for col in df.columns:
    plt.plot(df.index[-1], df[col].iloc[-1], 'o', color='navy')

# Styling
plt.title("Average Retail Product Price per Year", fontsize=14)
plt.xticks(df.index)
plt.yticks(fontsize=10)
plt.grid(False)
plt.tight_layout()

plt.show()
```
